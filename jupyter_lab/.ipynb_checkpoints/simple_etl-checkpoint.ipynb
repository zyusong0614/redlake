```python
from google.cloud import bigquery
```

```python
import os

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/Users/zhengyu/Desktop/UIUC/CS598/redlake/cert/sa1_key.json"
PROJECT_ID = "redlake-474918"
DATASET_ID = "redlake_dw"
TABLE_ID = "reddit_posts_raw"
URI = "gs://redlake/raw_json/posts/posts_2025-10-23_203844.json"
```

```python
# 初始化 BigQuery 客户端
bq = bigquery.Client(project=PROJECT_ID)

# 目标表 ID
table_id = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_ID}"

# 加载配置
job_config = bigquery.LoadJobConfig(
    source_format=bigquery.SourceFormat.NEWLINE_DELIMITED_JSON,
    autodetect=True,                     # 自动推断 schema
    write_disposition="WRITE_APPEND"     # 追加模式
)

# 启动加载任务
load_job = bq.load_table_from_uri(URI, table_id, job_config=job_config)
print(f"⏳ 正在加载 {URI} → {table_id}")
load_job.result()  # 等待完成

# 输出结果
table = bq.get_table(table_id)
print(f"✅ 加载完成: {table.num_rows} 行现存于 {table_id}")
```

```python
URI_COMMENTS = "gs://redlake/raw_json/comments/comments_2025-10-23_203844.json"
TABLE_COMMENTS = "reddit_comments_raw"

table_id_comments = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_COMMENTS}"

print(f"⏳ 正在加载 {URI_COMMENTS} → {table_id_comments}")
load_job = bq.load_table_from_uri(URI_COMMENTS, table_id_comments, job_config=job_config)
load_job.result()
print(f"✅ 加载完成: {bq.get_table(table_id_comments).num_rows} 行现存于 {table_id_comments}")
```
